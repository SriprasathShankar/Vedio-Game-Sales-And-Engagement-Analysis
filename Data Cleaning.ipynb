```python

column_to_shift = 'Unnamed: 0'

#Add 1 to every value in that column
games_copy['Unnamed: 0'] = games_copy['Unnamed: 0'] + 1

games.head(5)
```

```python
#Replace the first column with as crt serial number
games_copy.iloc[:, 0] = range(1, len(games_copy) + 1) 

#Rename the column name S.no
games_copy.rename(columns={'Unnamed: 0':'s_no'}, inplace=True)
```

```python
games_copy['Release Date'] = pd.to_datetime(games_copy['Release Date'], errors='coerce')
```

```python
games_copy.info()
```

```python
games_copy['Release Date'].isnull().sum()
```

```python
mode_data = games_copy['Release Date'].mode()[0]
games_copy['Release Date'].fillna(mode_data, inplace=True)
```

```python
games_copy['Release Date'].isnull().sum()
```

```python
# Converting the 3.9k into actual number

def convert_k(x):
    if isinstance(x, str) and 'K' in x: #Check if the value is a string and contains 'K'
        return float(x.replace('K','')) * 1000 #Remove 'K', convert to float (3.9), and multiply by 1000
        
    try: #Otherwise, try to convert the value directly to a float
        return float(x)
    except: #If conversion fails (e.g., 'N/A'), return NaN
        return np.nan
```

```python
#Converting Column values 3.9 to 3900
cols_to_convert = ['Times Listed', 'Number of Reviews', 'Plays', 'Playing', 'Backlogs', 'Wishlist']

for col in cols_to_convert:
    games_copy[col] = games_copy[col].apply(convert_k)
```

```python
cols = ['Team', 'Genres', 'Reviews']

for col in cols:
    games_copy[col] = games_copy[col].astype(str) #Make Sure Its String
    games_copy[col] = games_copy[col].replace(r"[\[\]\'\"\"]", "", regex=True) #Remove brackes and quotes
    games_copy[col] = games_copy[col].str.strip() #REmove lending and trailing spaces
```

```python
# Handle missing values: fill or drop
games_copy['Genres'] = games_copy['Genres'].fillna("Unknown")
games_copy['Team'] = games_copy['Team'].fillna("Unknown")
games_copy['Summary'] = games_copy['Summary'].fillna("No summary available")
games_copy['Reviews'] = games_copy['Reviews'].fillna("No reviews available")  # Fixed this line - changed games_copy() to fillna()
games_copy['Rating'] = games_copy['Rating'].fillna(games_copy['Rating'].mean())
games_copy['Times Listed'] = games_copy['Times Listed'].fillna(games_copy['Times Listed'].mean())
games_copy['Number of Reviews'] = games_copy['Number of Reviews'].fillna(games_copy['Number of Reviews'].mean())
```

```python
games_copy.isnull().sum()
```

```python
original_columns = games_copy.columns

new_columns = original_columns.str.lower().str.replace(' ','_')

games_copy.columns = new_columns
```

```python
#Filling missing values in the Year and publisher  column

most_common_year = sales_copy['Year'].mode()[0]
sales_copy['Year'] = sales_copy['Year'].fillna(most_common_year).astype(int)

sales_copy['Publisher'] = sales_copy['Publisher'].fillna("Unknown")
```

```python
# Convert categorical columns
categorical_cols = ['Name', 'Platform', 'Genre', 'Publisher']
for col in categorical_cols:
    sales_copy[col] = sales_copy[col].astype('category')
```

```python
#Setting Columns to lower and replace ' ' to ''
original_columns = sales_copy.columns
new_columns = original_columns.str.lower().str.replace(' ','_')

sales_copy.columns = new_columns
```
